```python
import matplotlib.pyplot as plt
```

```python
#create data to use matplotlib
import numpy as np
```

```python
delhi=np.array([12,14,16,32,64,56,11,43,35,41,34,28])
noida=np.array([32,42,21,52,43,24,39,32,64,56,11,43])
month=np.array(["jan","fab","mar","apr","may","jun","jul","aug","sep","oct","neb","dec"])
plt.ylabel("temp.")
plt.xlabel("mont.")
plt.plot(month,delhi,color="r")
plt.plot(month,noida,color="b")
```

# bar chart

```python
plt.figure(figsize=(10,9))
delhi=np.array([12,14,16,32,64,56,11,43,35,41,34,28])
noida=np.array([32,42,21,52,43,24,39,32,64,56,11,43])
month=np.array(["jan","fab","mar","apr","may","jun","jul","aug","sep","oct","neb","dec"])
plt.ylabel("temp.")
plt.xlabel("mont.")
plt.bar(month,delhi,color="r")
plt.bar(month,noida,color="g",align="edge")
```

# pie chart

```python
# when a numerical data is given accros a catagorical data than we use pie chart
a=[12,43,56,34,23]
item=["A","B","C","D","E"]
plt.pie(a,labels=item,autopct="%0.2f%%")
```
