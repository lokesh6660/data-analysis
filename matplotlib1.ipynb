```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
mon=np.array(["jan","fab","mar","apr","may","jun","jul","aug","sep","oct","neb","dec"])
delhi=np.array([34,36,32,35,23,45,27,16,23,45,32,42])
noida=np.array([31,32,42,55,23,35,37,26,43,35,22,32])
plt.xlabel("mon.")
plt.xlabel("temp.")
plt.plot(mon,delhi,color="r") 
plt.plot(mon,noida,color="b") 
```

```python
#numerical/numerical
x=np.array([23,34,12,46])
y=np.array([1,2,3,4])
plt.plot(x,y)
```

```python
#numerical/categorical
x=np.array([23,34,12,46])
y=np.array(["A","B","C","D"])
plt.plot(x,y)
```

```python
#categorical/categorical
x=np.array(["W","X","Y","Z"])
y=np.array(["A","B","C","D"])
plt.plot(x,y)
```

```python
#categorical/categorical
x=np.array([True,False,True])
y=np.array(["A","B","C"])
plt.plot(y,x)
#T-----> 1
#F-----> 0
```
