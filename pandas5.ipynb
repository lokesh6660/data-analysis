```python
#in founded and headquarters coloum only one data missing so we ignore it because it do
#not create any change in analysis.
```

```python
#percent of missing data in Founder/s
a=df["Founder/s"].isnull().sum()
b=df.shape[0]
per=(a/b)*100
print("percent of missing data in Founder/s= ",per)
```

```python
#percent of missing data in Founded
a=df["Founded"].isnull().sum()
b=df.shape[0]
per=(a/b)*100
print("percent of missing data in Founded= ",per)
```

```python
#percent of missing data in Investor/s
a=df["Investor/s"].isnull().sum()
b=df.shape[0]
per=(a/b)*100
print("percent of missing data in Investor/s = ",per)
```

```python
#percent of missing data in Amount
a=df["Amount"].isnull().sum()
b=df.shape[0]
per=(a/b)*100
print("percent of missing data in Amount = ",per)
```

```python
#percent of missing data in Stage
a=df["Stage"].isnull().sum()
b=df.shape[0]
per=(a/b)*100
print("percent of missing data in Stage= ",per)
```

```python
#percent of total missing data in data set
a=df.isnull().sum().sum()
b=df.shape[0]
per=(a/b)*100
print("percent of total missing data in data set= ",per)
```
