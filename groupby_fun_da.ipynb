```python
df=pd.read_csv(r"C:\Users\hp\Desktop\IPL Matches 2008-2020.csv")
df
```

```python
#use median with groupby function




df[df["city"]=="Delhi"].groupby(["toss_winner"]).median()
```

```python
#use median with groupby function



df[df["city"]=="Delhi"].groupby(["toss_winner"]).mean()
```

```python
# to find which team win toss with its player of match in group


df.groupby(["toss_winner","player_of_match"]).mean()
```

```python
#we can use "iloc" function with groupby to probide the variable lenth range of the 
#data in the data set.



df.groupby(["toss_winner","player_of_match"]).mean().iloc[0:5]
```
