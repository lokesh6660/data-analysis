```python
import numpy as np
import matplotlib.pyplot as plt
```

# barchart----->

```python

plt.figure(figsize=(10,9))
mon=np.array(["jan","fab","mar","apr","may","jun","jul","aug","sep","oct","nov","dec"])
delhi=np.array([34,36,32,35,23,45,27,16,23,45,32,42])
noida=np.array([31,32,42,55,23,35,37,26,43,35,22,32])
plt.xlabel("mon.")

plt.xlabel("temp.")
plt.bar(mon,delhi,color="r",align="edge") 
plt.bar(mon,noida,color="b") 
```

# pie chart ---------->


```python
#pie chart work on categ. data 
a=[45,67,34,29]
myLabel=["A","B","C","D"]

plt.pie(a,labels=myLabel,autopct="%1.0f%%")
```
